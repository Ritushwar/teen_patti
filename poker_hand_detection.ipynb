# POKER HAND DETECTION

## IMPORTING REQUIRED LIBRARIES

```python
from ultralytics import YOLO
import cv2
import cvzone
import math
import csv
import torch
import numpy as np
import teen_patti
```

## IMPORTING MODEL

```python
model = YOLO("/home/ritu/object_detection/POKER HAND DETECTION/playingCards.pt")
```

## IDENTIFYING TOTAL CLASSES

```python
total_classes = []
classes = model.names
#print(classes)
for clss in classes:
    #print(classes[clss])
    total_classes.append(classes[clss])
print(len(total_classes))
print(total_classes)
```

## PASSING IT TO MODEL

```python
cap = cv2.VideoCapture("/home/ritu/object_detection/POKER HAND DETECTION/sample_vdo2.mp4")
# cap.set(3,640)
# cap.set(4,480)
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))
frame_fps = int(cap.get(5))
results_dict = {6:"Trail",5:"Doubly",
                    4:"Run",3:"Color",
                    2:"Juut",1:"High Card"}
while True:
    success, img = cap.read()
    if not success:
        break
    results = model(img)
    pokers = []
    for r in results:
        pokers = []
        boxes = r.boxes
        for box in boxes:
            # bounding box
            x1,y1,x2,y2 = box.xyxy[0]
            x1,y1,x2,y2 = int(x1),int(y1),int(x2),int(y2)
            conf = math.ceil((box.conf[0]*100)) / 100
            clss = int(box.cls[0])
            if conf > 0.4:
                current_class = total_classes[clss]
                if current_class not in pokers:
                    pokers.append(current_class)
                w,h = x2-x1, y2-y1
                cvzone.cornerRect(img, (x1,y1,w,h),colorC=(0,0,255),l=5)
                cvzone.putTextRect(img, f'{current_class}',(max(0,x1),max(0,y1)), scale=0.6, thickness=1)
    if len(pokers)==5:
        possibleRanks = int(find.findPokerHand(pokers)[0])
        pokers=[]
        cvzone.putTextRect(img, f'{results_dict[possibleRanks]}',(max(0,10),max(0,20)), scale=2, thickness=1)
    cv2.imshow("Result",img)
    key = cv2.waitKey(0) & 0xFF
    if key == ord('q'):
        break
    if key == 27:
        break 
cap.release()
cv2.destroyAllWindows()
```
